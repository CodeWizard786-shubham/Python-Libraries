1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import numpy as np

try:
    x_axis = np.random.randint(100, size=1000)
    y_axis = np.random.randint(100, size=1000)
    fig = px.scatter(x=x_axis, y=y_axis, title='Scatter Plot of Random 1000 Coordinates')
    fig.show()

except Exception as e:
    logger.error(str(e))
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

try:
    x_axis = np.random.randint(100, size=100)
    y_axis = np.random.randint(100, size=100)

    fig = px.line(x=x_axis,y=y_axis , title='Line and scatter Plot of Random Coordinates',width=800, height=400)
    scatter_fig = px.scatter(x=x_axis, y=y_axis)

    fig3 = go.Figure(data = fig.data + scatter_fig.data)

    fig3.update_layout(autosize=False)
    fig3.show()

    logger.info("Plotly scatter plot for random 100 x and y coordinates")

except Exception as e:
    logger.error(str(e))
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
import plotly.express as px
import numpy as np

try:
    x_axis = np.random.randint(100,size=(500))
    y_axis = np.random.randint(100,size=(500))

    fig = px.scatter(x=x_axis , y =y_axis,title="Scatter plot for random 500 coordinates",opacity=0.8,color=y_axis)

    fig.update_layout(
    margin=dict(l=80, r=80, t=40, b=20),
    paper_bgcolor="LightSteelBlue",
    autosize =False
    )
    fig.update_traces(marker=dict(size=12,line=dict(width=2,color='red')),
                      selector=dict(mode='markers')
                      )
    
    fig.show()

    logger.info("Plotly Scatter plot for 500 coordinates")

except Exception as e:
    logger.error(str(e))
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

try:
    df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv')    

    print(df.head())
    fig = go.Figure(px.scatter(df,x='Population',y='State',hover_data='State',title='USA States (2014)'))

    fig.update_layout(autosize=False)
    fig.show()

    logger.info("Ploty scatter plot ")

except Exception as e:
    logger.error(str(e))
```
