9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.


```python
import matplotlib.pyplot as plt
import pandas as pd

try:
    Programming_languages = pd.Series(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    Popularity = pd.Series([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    # bar chart with uniform color
    fig = plt.bar(Programming_languages,Popularity,color=['y', 'r', 'g', 'b', 'c','0'],edgecolor = 'blue')

    plt.subplots_adjust(bottom=0.7)
    plt.xlabel('Programming languages')
    plt.ylabel('Popularity')
    
    #label bars with values
    plt.bar_label(fig, label_type='edge')

    plt.show()

    logger.info("Bar chart with specific size for each bar plotted sucessfully ")


except Exception as e:
    logger.error(str(e))
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.


```python
import matplotlib.pyplot as plt
import numpy as np

try:
    groups = ['Group1', 'Group2', 'Group3', 'Group4', 'Group5', 'Group6']
    men_scores = [22, 30, 35, 35, 26, 30]
    women_scores = [25, 32, 30, 35, 29, 32]

    # Set the width of the bars
    bar_width = 0.35

    # Calculate the X positions for the bars
    x = np.arange(len(groups))

    # Create the bar plot
    plt.bar(x, men_scores, bar_width, label='Men')
    plt.bar(x + bar_width, women_scores, bar_width, label='Women')

    # Set the labels and title
    plt.xlabel('Groups')
    plt.ylabel('Scores')
    plt.title('Scores by Group and Gender')

    # Set the X-axis ticks and labels
    plt.xticks(x + bar_width/2,groups)

    plt.legend()

    plt.show()
except Exception as e:
    
    logger.error(str(e))
```

11. Write a Python program to create bar plot from a DataFrame.

```python
import matplotlib.pyplot as plt
import pandas as pd

try:
    # Create the DataFrame
    data = {'Group':['A','B','C','D','E'],
            'a': [2, 4, 6, 8, 10],
            'b': [4, 2, 4, 2, 4],
            'c': [8, 5, 4, 4, 3],
            'd': [5, 7, 4, 8, 3],
            'e': [7, 6, 7, 6, 2]}
    df = pd.DataFrame(data)

    # Create the bar plot
    df.plot(x='Group',kind='bar')

    # Set the labels and title
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Bar Plot from DataFrame')
    # Show the plot
    plt.show()

    logger.info("Bar chart plotted from dataframe")
except Exception as e:
    logger.error(str(e))
```

12. Write a Python program to create bar plots with error bars on the same figure.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

try:
    x = ["A","B","C","D"]

    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]

    plt.bar(x,mean_velocity,yerr = std_deviation,color='lightgray',edgecolor='black',ecolor='red',capsize =5)    

    plt.show()

    logger.info("Bar chart with error bars for each bar plotted sucessfully")
except Exception as e:
    logger.error(str(e))
```
