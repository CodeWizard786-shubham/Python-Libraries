5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python

import matplotlib.pyplot as plt
import numpy as np
  
# line 1
x = np.array([2, 4, 6, 8])
y = np.array([2, 4, 6, 8])
# first plot with X and Y data
plt.plot(x, y,label='line1')
  
# line 2
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
# second plot with x1 and y1 data
plt.plot(x1, y1,label='line2')
  
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.grid()
plt.legend()
plt.show()
         
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

try:
    # Read data from csv file using pandas
    df = pd.read_csv('fdata.csv', parse_dates=['Date'], index_col=0)

    # Plot the stock prices with line styles and markers
    plt.plot(df['Open'], label='Open', linestyle='-', linewidth=2, marker='>')
    plt.plot(df['High'], label='High', linestyle='--', linewidth=3, marker='o')
    plt.plot(df['Low'], label='Low', linestyle='-.', linewidth=4, marker='P')
    plt.plot(df['Close'], label='Close', linestyle=':', linewidth=5, marker='D')

    # Customize the graph
    plt.xticks(rotation=45, color='orange')
    plt.yticks(color='orange')
    plt.xlabel('Date', color='red')
    plt.ylabel('Stock Prices', color='red')
    plt.title('Financial Data (Alphabet Inc.)')
    plt.grid(True)
    plt.legend(loc='upper left')

    # Find current axis limits
    current_xlim = plt.xlim()
    current_ylim = plt.ylim()
    print(f"Current X-axis limits: {current_xlim}")
    print(f"Current Y-axis limits: {current_ylim}")

    # Set new axis limits with date values
    start_date = datetime.strptime('15-04-16', '%d-%m-%y')
    end_date = datetime.strptime('15-06-16', '%d-%m-%y')
    new_xlim = (start_date, end_date)
    new_ylim = (775, 780)
    plt.xlim(new_xlim)
    plt.ylim(new_ylim)

    # Display the updated axis limits
    print(f"New X-axis limits: {plt.xlim()}")
    print(f"New Y-axis limits:{plt.ylim()}")

    plt.show()
    logger.info("Read data from csv and set new limits to axis on line successfully")

except Exception as e:
    logger.error(str(e))
```

14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.


```python
import matplotlib.pyplot as plt

try:
    
    dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
    closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

    plt.plot(dates, closing_values, marker='o')

    # Customize major grid lines
    plt.grid(which='major',linestyle='-', linewidth=0.8, color='red')
    # Customize minor grid lines
    plt.grid(which='minor',linestyle=':', linewidth=0.5, color='blue')

    # Show the minor ticks and grid.
    plt.minorticks_on()
    # Hide the minor ticks but show gridlines.
    plt.tick_params(which='minor', bottom=False, left=False)

    # Set labels and title
    plt.xlabel('Date')
    plt.ylabel('Closing Value')
    plt.title('Closing Value of Alphabet Inc.')

    # Rotate x-axis 
    plt.xticks(rotation=45)

    # Display the plot
    plt.show()

except Exception as e:
    logger.error(str(e))
```

15. Write a Python program to create multiple plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Read the CSV file into a DataFrame
    df = pd.read_csv('fdata.csv', parse_dates=['Date'], index_col=0,date_format='%d-%m-%y')

    # Extract the necessary columns
    dates = df.index
    open = df['Open']
    high = df['High']
    low = df['Low']
    close = df['Close']

    # Create subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

    ax1.plot(dates, open)
    ax1.set_title('Open Values')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Open')
    ax1.grid()
    ax1.tick_params(axis='x', rotation=45)
    


    ax2.plot(dates, high)
    ax2.set_title('High Values')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('High')
    ax2.grid()
    ax2.tick_params(axis='x', rotation=45)


    ax3.plot(dates, low)
    ax3.set_title('Low Values')
    ax3.set_xlabel('Date')
    ax3.set_ylabel('Low')
    ax3.grid()
    ax3.tick_params(axis='x', rotation=45)


    ax4.plot(dates, close)
    ax4.set_title('Close Values')
    ax4.set_xlabel('Date')
    ax4.set_ylabel('Close')
    ax4.grid()
    ax4.tick_params(axis='x', rotation=45)


    # spacing between subplots
    plt.tight_layout()

    # Display the plots
    plt.show()

except Exception as e:
    print(str(e))
```
