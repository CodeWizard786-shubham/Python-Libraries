1. Write a Python programming to create a pie chart of the popularity of programming Languages.


```python
import matplotlib.pyplot as plt
import pandas as pd

try:
    Programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    fig, ax = plt.subplots()

    # plotting pie chart of polularity and labels are programming languages.
    ax.pie(Popularity, labels=Programming_languages)

    plt.show()

    logger.info("Pie chart created successfully")

except Exception as e:
    logger.error(str(e))
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.

```python
import matplotlib.pyplot as plt

try:
    Programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    fig,ax = plt.subplots()

    ax.pie(Popularity, labels=Programming_languages,autopct="%1.1f%%")

    plt.title('Popularity of Programming languages')
    plt.show()

    logger.info("Set title and pct to pie chart")


except Exception as e:
    logger.error(str(e))
```

3. Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
import matplotlib.pyplot as plt
import pandas as pd

try:
    # Read data from csv file
    df = pd.read_csv('medal.csv')

    fig , ax = plt.subplots()

    country_values = df['country']

    gold_medal_count = df['gold_medal']

    # plot pie chart of gold medals and labels as country
    patches,texts,autotexts = ax.pie(gold_medal_count, labels=country_values,autopct = '%1.1f%%',shadow=True,
           startangle=180,wedgeprops={'linewidth':2,'edgecolor':"white",'antialiased': True})
    
    [autotext.set_color('blue') for autotext in autotexts]

    for i ,patch in enumerate(patches):
        texts[i].set_color(patch.get_facecolor())


    # customise pie chart
    plt.title("2016 Summer Olympics",color='green')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

    logger.info("Applied different properties to pie chart")

except Exception as e:
    logger.error(str(e))


```
