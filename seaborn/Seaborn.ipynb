1. Write a program to draw bar plot of sex against survived for a dataset given in the url


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

try:
    # Load the Titanic dataset from the CSV file
    df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")

    # Display the first few rows of the dataset

    logger.info("Titianic Dataset sucessfully loaded")
    print(df.head(10))

    # plot seabrom bar plot
    #sns.barplot(x=df['sex'] , y=df['survived'])

    fig = sns.barplot(x='sex' , y='survived' , data=df , label=['male','female'])

    plt.title('Titanic Sex vs Survived seaborn Bar Plot')

    plt.legend(title='Sex')

    plt.show()

    logger.info("Seaborn bar plot plotted successfully")

except Exception as e:
    logger.error(str(e))
```

4. Write a program to draw a violin plot of sex against total_bill for a given dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

try:
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

    print(df.head(10))

    # plotting seaborn violin plot
    sns.violinplot(x='sex', y='total_bill' , data=df ,hue='sex')

    plt.title('Total bill by sex Violin plot')

    plt.show()
    logger.info("Seaborn violin plot successfully plotted")

except Exception as e:

    logger.error(str(e))
```

6. Write a program to draw box plot of life expectancy by continent for a data set given in a url

7. Write a program to draw a box plot of day by tips for a dataset given in a url

```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

try:

    
    url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'

    tips_data = pd.read_csv(url)

    print(tips_data.head(10))

    # seaborn box plot
    sns.boxplot(x = 'day',y='tip' ,data = tips_data, width=0.8 , hue='sex' ,order=['Thur','Fri','Sat','Sun'])

    plt.title("Day by tips") 
    
    plt.show()
    
    logger.info("Seaborn box plot for tips dataset")


except Exception as e:

    logger.error(str(e))
```

8. Write a program to draw a swarm plot of total bill against size for a given dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

try:

    tips_data = sns.load_dataset('tips')

    sns.swarmplot(x='size' , y='total_bill' , data = tips_data , hue='sex')

    plt.title("Total bill by size")
    plt.show()

    logger.info('seaborn Swam Plotted successfully')

except Exception as e:

    logger.error(str(e))

```
